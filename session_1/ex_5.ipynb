```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Create a dictionary with data
data = {
    "Year": range(1970, 1985),
    "Wells": [28, 26, 28, 28, 33, 39, 41, 46, 50, 70, 52, 90, 84, 74, 82],
}

# Create DataFrame
df = pd.DataFrame(data)

# Set the index to be the Year column
df.set_index("Year", inplace=True)
```

```python
df["Wells"].plot(
    kind="line",
    marker="o",
    figsize=(10, 6),
    legend=False,
    title="Number of Wells for each Year",
)

plt.grid(True)
plt.show()
```

The time series data has a increasing trend.

```python
# Calculate moving average with window of 3 years
df["MA"] = df["Wells"].rolling(window=4).mean()

df["MA"]
```

```python
beta_1, beta_0 = 6.461246201, 49.8612462

y_pred = beta_0 + beta_1 * df.index
```

```python
plt.plot(df.index, df["Wells"], "-o")
plt.plot(df.index, y_pred, "-")
```
