```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a dictionary with data
data = {
    "Year": [1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999],
    "Y(t)": [7, 8, 10, 12, 14, 16, 25, 28, 30],
    "MA": [None, 8.33, 10, 12, 14, 18.33, 23, 27.67, None],
    "ES": [None, 7, 7.4, 8.44, 9.864, 11.5184, 13.311, 17.9866, None],
}

# Create a dataframe from the dictionary
df = pd.DataFrame(data)

# Add detrend columns to the dataframe
df["detrend(ma)"] = df["Y(t)"] - df["MA"]
df["detrend(es)"] = df["Y(t)"] - df["ES"]

# Set the year column as the DataFrame's index
df = df.set_index("Year")

# Create a line plots
df["Y(t)"].plot(kind="line", label="Y(t)", marker="o", alpha=0.5)
df["MA"].plot(
    kind="line", label="Moving Average", marker="o", color="red", linestyle="dashed"
)
df["ES"].plot(
    kind="line",
    label="Exponential Smoothing",
    marker="o",
    color="green",
    linestyle="dashed",
)
df["detrend(ma)"].plot(
    kind="line", label="Detrend(MA)", marker="x", color="orange", linestyle="dotted"
)
df["detrend(es)"].plot(
    kind="line", label="Detrend(ES)", marker="x", color="purple", linestyle="dotted"
)

# Add a title and labels
plt.title("Data Trends Over the Years")
plt.ylabel("Values")
plt.grid()
plt.legend()

plt.show()
```
